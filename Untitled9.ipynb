## Q3. Whi,h Python statements are used to ,catch and handle the exceptions? Explain with an example


```python
 Python statements used to catch and handle exceptions. In Python, try, except, else, and finally are used to catch and handle exceptions. Here's an explanation along with an example:

try: This block of code is used to enclose the statements that might raise an exception.

except: This block of code is executed if an exception occurs within the corresponding try block. You can specify which type of exceptions to catch, or use a generic except to catch all exceptions.

else: This block of code is executed if no exceptions are raised in the try block.

finally: This block of code is always executed, regardless of whether an exception occurred or not. It's typically used to perform cleanup actions, such as closing files or releasing resources.

Here's an example to illustrate the usage:

def divide_numbers(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Error: Division by zero!")
    except TypeError:
        print("Error: Unsupported operand type!")
    else:
        print("Division successful. Result:", result)
    finally:
        print("This block always executes, regardless of exceptions.")

# Example usage:
divide_numbers(10, 0)  # Division by zero
print("------------------")
divide_numbers(10, 'a')  # Unsupported operand type
In this example:

The try block attempts to perform a division operation.
The first except block catches the ZeroDivisionError if y is zero.
The second except block catches TypeError if the input types are incompatible.
The else block executes if no exceptions occurred, printing the result.
The finally block always executes, regardless of whether an exception occurred or not, performing cleanup actions.
This comprehensive error handling ensures that the program gracefully handles exceptions and provides appropriate feedback to the user.





```
